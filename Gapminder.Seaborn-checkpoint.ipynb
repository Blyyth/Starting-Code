```python

plt.plot(us.year, us.lifeExp)
plt.plot(afg.year, afg.lifeExp)
plt.title("Comparing U.S and Afghanistan life Expectancy by year")
plt.xlabel("Year")
plt.ylabel("Life Exp")

plt.show()
```

```python
## creating a grpah using seaborn and pandas to see the
## average life expectancy per country

country_list = list(gapminder.country.unique())
avg_life = []

for i in country_list:
    x = gapminder[gapminder.country == i]
    country_avg = sum(x.lifeExp)/len(x)
    avg_life.append(country_avg)
    
df1 = pd.DataFrame({'country_list':country_list,'avg_life':avg_life})

plt.figure(figsize = (15,10))
sns_graph = sns.barplot( x = df1.country_list, y = df1.avg_life)


plt.xticks(rotation = 90)
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy')
plt.title('Average Life Expectancy per country')
sns_graph.tick_params(axis='x', labelsize=5)
```
